```python
import matplotlib.pylab as plt
```

```python
velika_noc = {}
for leto in range(1582, 99999999):
    k = leto // 100
    p = (8 * k + 13) // 25
    q = k // 4
    m = 15 + k - p - q
    n = 4 + k - q
    a = leto % 19
    b = leto % 4
    c = leto % 7
    d = (19 * a + m) % 30
    e = (2 * b + 4 * c + 6 * d + n) % 7
    datum = 22 + d + e
    velika_noc[datum] = velika_noc.get(datum, 0) + 1
```

```python
courses = list(velika_noc.keys())
values = list(velika_noc.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("Datum velike noči (kateri marec)")
plt.ylabel("Število pojavitev")
plt.title("Pogostost datumov velike noči")
plt.show()
```

```python
lists = sorted(velika_noc.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()
```
